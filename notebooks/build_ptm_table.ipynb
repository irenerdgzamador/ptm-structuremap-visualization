# PTM Table Construction — Preprocessing Notebook

This notebook performs the preprocessing of raw peptide modification data
and generates a binary PTM table compatible with the downstream analysis
software.

All heavy processing logic is implemented in `src/utils.py`.  
This notebook only:

1. Loads the configuration from `config.yaml`
2. Calls the preprocessing pipeline
3. Optionally displays a preview of the processed files

This ensures a clean, reproducible and modular workflow.


```python
#Imports
import yaml
import pandas as pd
from src.utils_build_ptm_table import process_all_files
```

```python
# Load configuration file
with open("../config.yaml", "r") as f:
    config = yaml.safe_load(f)

input_folder = config["input_folder"]
output_folder = config["output_folder"]
separator = config["separator"]

print("Configuration loaded:")
print(f"  Input folder: {input_folder}")
print(f"  Output folder: {output_folder}")
print(f"  Separator: '{separator}'")
```

```python
#Run preprocessing pipeline
process_all_files(
    input_folder=input_folder,
    output_folder=output_folder,
    sep=separator
)
```

```python
#Optional: preview processed files
import glob
import os

processed_files = glob.glob(os.path.join(output_folder, "*"))

print(f"Processed files found: {len(processed_files)}")

if processed_files:
    df_preview = pd.read_csv(processed_files[0])
    print("\nPreview of the first processed file:")
    display(df_preview.head())
```
